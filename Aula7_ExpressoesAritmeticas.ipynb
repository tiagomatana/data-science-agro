### Expressões aritméticas

* Soma: `+`
* Subtração: `-`
* Multiplicação: `*`
* Divisão: `/`
* Divisão inteira: `//`
* Resto da divisão: `%`
* Exponenciação: `**`

```python
2**5
```

- *strings* (str) suportam os operadores ``+`` (concatenação) e ``*`` (replicação), resultando em novas strings.

```python
# Exemplo da "aritmética" com strings
y = 'Aprendendo'
y = y + ' Python '
print("Concatenacao y + ' Python ': ",y)
w = 3*y   # A operação de replicação "*" asume um número inteiro e uma string como argumentos
print("3*y replica y três vezes: ",w)
#Verificar se um texto está dentro do outro
print('agron' in 'agronegocio')
```

#### Conversão: explícita e implícita

```python
# exemplos de conversão explícita e implícita
x = 2.76
y = int(x)  # convertendo float para int
z = y == 2  # resultado da comparação y == 1 (True/False) é armazenado em z

print("x =", x, " | tipo:", type(x))
print("y =", y, "    | tipo:", type(y))
print("z =", z, "| tipo:", type(z))
```

### Método format

```python
faturamento_vendas = 2000.25
custo_adubo = 500
lucro = faturamento_vendas - custo_adubo

print('O faturamento da loja foi ' + str(faturamento_vendas) + '.O custo da loja foi ' + str(custo_adubo) + '.Assim, o lucro da loja foi de ' + str(lucro))
```

```python
faturamento_vendas = 2000.252541
custo_adubo = 500
lucro = faturamento_vendas - custo_adubo

print('O faturamento da loja foi {:.3f}. O custo da loja foi {}. Assim, o lucro da loja foi {}'.format(faturamento_vendas, custo_adubo, lucro))
```
